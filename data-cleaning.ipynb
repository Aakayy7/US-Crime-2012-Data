```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("U.S._Offense_Type_by_Agency_2012.csv")

df
```

```python
deleted = 1
df = df.drop(deleted)
df
```

```python
values = df.loc[2,['United States', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4',
       'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8']]

othervalues =  df.loc[3,['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12',
       'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16',
       'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20',
       'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24',
       'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 'Unnamed: 28',
       'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32',
       'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35', 'Unnamed: 36',
       'Unnamed: 37', 'Unnamed: 38', 'Unnamed: 39', 'Unnamed: 40',
       'Unnamed: 41', 'Unnamed: 42', 'Unnamed: 43', 'Unnamed: 44',
       'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47', 'Unnamed: 48',
       'Unnamed: 49', 'Unnamed: 50', 'Unnamed: 51', 'Unnamed: 52',
       'Unnamed: 53', 'Unnamed: 54', 'Unnamed: 55', 'Unnamed: 56',
       'Unnamed: 57', 'Unnamed: 58', 'Unnamed: 59', 'Unnamed: 60','Unnamed: 61','Unnamed: 62']]

value_list = values.tolist()
other_list = othervalues.tolist()

```

```python
value_list[4] = "Total Offenses"
value_list[5] = "Crimes Against Persons"
value_list[6] = "Crimes Against Property"
value_list[7] = "Crimes Against Society"
```

```python
other_list
```

```python
columns_list = value_list + other_list 
columns_list
```

```python
new_list = []

for item in columns_list :
    new_list.append(item.replace('\n', ' ')) # delete \n
    
print(new_list)

df.columns = new_list

df.head(10)
```

```python
data = df.copy()

data.drop(data.tail(4).index,inplace=True)
```

```python
data.shape
```

```python
data = data.drop([0, 1]) # delete 0 and 1 indexes
data = data.reset_index(drop=True)

data
```

```python
data.drop(["Agency Type"], axis = 1 ) 
```

```python
data.rename(columns={"Agency Name": "City"}, inplace=True) 
```

```python
data[['State', 'Agency Type']] = data[['State', 'Agency Type']].fillna(method='ffill') # fill the empty states and citys

data
```

```python
data.info()
```

```python
drop_list = []

# For loop to calculate the percentage of null values in all columns
for column in data.columns:
    total_null = (data[column].isna().sum()) / len(data) * 100
    print(column + ': ' + '%' + str(total_null))
    
    if (total_null > 70).any():
        drop_list.append(column)

# Now 'drop_list' contains the names of columns with more than 70% null values
print("Columns to drop:", drop_list)
```

```python
for drop_column in drop_list:
    
    data.drop(drop_column, inplace=True, axis=1)
   
data.head()
```

```python
deleted = ["Total Offenses"]
data = data.drop(columns = deleted)
data
```

```python
# columns that will be deleted

to_delete = ['Drug Equipment Violations','Porno- graphy/ Obscene Material',
            'Theft of Motor Vehicle Parts or Accessories', 'All Other Larceny',
            'Drug/ Narcotic Offenses','Theft From Motor Vehicle', 'Larceny- Theft Offenses',
             'Shop- lifting','Imperson- ation','Embezzle- ment', 'False Pretenses/ Swindle/ Confidence Game',
            'Counter- feiting/ Forgery','Arson','Rape','Fondling','Intimidation','Aggravated Assault', 'Simple Assault']

data = data.drop(columns = to_delete)

data.head()
```

```python
data.columns
```

```python
columns_to_process = ['Population1', 'Crimes Against Persons', 'Crimes Against Property', 'Crimes Against Society',
                      'Kidnapping/ Abduction', 'Sex Offenses', 'Sex Offenses, Nonforcible',
                      'Burglary/ Breaking & Entering', 'Destruction/ Damage/ Vandalism', 'Fraud Offenses',
                      'Credit Card/ Automated Teller Machine Fraud', 'Theft From Building', 'Motor Vehicle Theft',
                      'Robbery', 'Stolen Property Offenses ', 'Drug/ Narcotic Violations', 'Weapon Law Violations']


for column in columns_to_process:
    # Use a lambda function to conditionally convert values to integers
    data[column] = data[column].apply(lambda x: int(x.replace(',', '')) if isinstance(x, str) else x)

    
data
```

```python
pa_li = ['Population1', 'Crimes Against Persons',
       'Crimes Against Property', 'Crimes Against Society',
       'Crimes Against Persons', 'Kidnapping/ Abduction', 'Sex Offenses',
       'Sex Offenses, Nonforcible', 'Burglary/ Breaking & Entering',
       'Destruction/ Damage/ Vandalism', 'Fraud Offenses',
       'Credit Card/ Automated Teller Machine Fraud', 'Theft From Building',
       'Motor Vehicle Theft', 'Robbery', 'Stolen Property Offenses ',
       'Drug/ Narcotic Violations', 'Weapon Law Violations']


# Calculate the median for each column in 'pa_li'

for pa in pa_li:
    valid_columns = data.select_dtypes(include=[int, float]).columns
    if pa in valid_columns:
        pa_median = data[pa].median()
        data[pa] = data[pa].fillna(pa_median).copy()


# fill the empty cells with median
```
